# Análise exploratória de dados

## Qual o valor médio de aluguel por tipo de imóvel?

```python
dados.head()
```

```python
# Calcular a média da coluna 'Valor'
dados['Valor'].mean()
```

```python
# Calcular a média da coluna 'Valor' agrupada por 'Tipo', calculando apenas valores numéricos
dados.groupby('Tipo').mean(numeric_only=True)
```

```python
# Calcular a média da coluna 'Valor' agrupada por 'Tipo'
dados.groupby('Tipo')['Valor'].mean()
```

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# Visualizar a média de preço por tipo de imóvel em um gráfico de barras horizontais
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh',figsize=(14,10), color='purple', title='Média de Preço por Tipo de Imóvel')
```

## Removendo os imóveis comerciais

```python
# Listar os tipos únicos de imóveis na coluna 'Tipo'
dados.Tipo.unique()
```

```python
# Filtrar o DataFrame para incluir apenas imóveis comerciais
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
# removendo imóveis comerciais
df_comerciais = dados.query('@imoveis_comerciais in Tipo')
df_comerciais.head()
```

```python
# Filtrar o DataFrame para incluir apenas imóveis residenciais
df_residenciais = dados.query('@imoveis_comerciais not in Tipo')
df_residenciais.head()
```

```python
# Visualizar a média de preço por tipo de imóvel em um gráfico de barras horizontais
df_residenciais.Tipo.unique()
```

```python
# Visualizar a média de preço por tipo de imóvel em um gráfico de barras horizontais
df_preco_tipo = df_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh',figsize=(14,10), color='purple', title='Média de Preço por Tipo de Imóvel')
```
