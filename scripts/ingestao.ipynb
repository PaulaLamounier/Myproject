```python
import duckdb
import pandas as pd
import os
from datetime import datetime
```

```python
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
arquivo = 'z0019_2.csv'
data_ingestao = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
df = pd.read_csv(f'../landing/{arquivo}', sep=';')
df['nome_arquivo'] = arquivo
df['data_ingestao'] = data_ingestao
df.head()
```

```python
#ingestão dos arquivos na tabela bronze 
con .execute("""
    CREATE TABLE IF NOT EXISTS bronze_produtos (
        NARBR VARCHAR,
        MAKTX VARCHAR,
        WERKS VARCHAR,
        MAINS VARCHAR,
        LABST VARCHAR,
        nome_arquivo VARCHAR,
        data_ingestao TIMESTAMP)
""")
```

```python
#Inserção dos dados do dataframe na tabela bronze
con.execute("Insert into bronze_produtos select * from df")
```

```python
#validar conteúdo da tabela bronze
resultado = con.execute("select * from bronze_produtos").fetchdf()
print(resultado)
```
